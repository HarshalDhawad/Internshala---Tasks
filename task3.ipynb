```python
import pandas as pd
```

```python
raw_data_path = 'rawdata.xlsx'
raw_data = pd.read_excel(raw_data_path)
raw_data
```

```python
raw_data['date'] = raw_data['date'].astype(str)
raw_data['time'] = raw_data['time'].astype(str)
```

```python
raw_data['datetime'] = pd.to_datetime(raw_data['date'] + ' ' + raw_data['time'])
raw_data['date'] = raw_data['datetime'].dt.date
```

```python
raw_data
```

```python
raw_data['duration'] = raw_data['datetime'].diff().dt.total_seconds().fillna(0)
raw_data
```

```python
raw_data['activity'] = raw_data.apply(lambda row: 'picked' if row['sensor'] == 1 else 'placed', axis=1)
raw_data
```

```python
# Define a function to classify locations
def classify_location(location):
    if location.startswith('A'):
        return 'inside'
    else:
        return 'outside'

raw_data['location_type'] = raw_data['location'].apply(classify_location)

# Group by date and location type to calculate total duration
duration_summary = raw_data.groupby(['date', 'location_type'])['duration'].sum().reset_index()
duration_summary = duration_summary.pivot(index='date', columns='location_type', values='duration').fillna(0).reset_index()
duration_summary.columns.name = None  # Remove the pivot table column name

# Display the duration summary
print(duration_summary)
```

```python
# Group by date and activity to count the number of activities
activity_summary = raw_data.groupby(['date', 'activity'])['activity'].count().unstack().fillna(0).reset_index()

# Display the activity summary
print(activity_summary)
```

```python
# Merge the duration and activity summaries
final_summary = pd.merge(duration_summary, activity_summary, on='date', how='outer').fillna(0)

# Display the final summary
print(final_summary)

# Save the final summary to an Excel file
final_summary_path = 'final_summary.xlsx'
final_summary.to_excel(final_summary_path, index=False)
```
